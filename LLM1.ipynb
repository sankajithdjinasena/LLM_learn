```python
# Import necessary Libraries
import os
from google.colab import userdata
```

```python
from langchain_openai import ChatOpenAI
os.environ["OPENAI_API_KEY"] = userdata.get('OPENAI_API')

llm = ChatOpenAI(
    model_name="gpt-3.5-turbo", # OPEN AI model
    temperature=0
)
```

### Initialize Prompt Template

```python
from langchain_core.prompts import ChatPromptTemplate

# Create a prompt template
prompt = ChatPromptTemplate.from_messages(
    [
        ("system","You are an intelligent chatbot. Answer the follwong question"),
        ("user","{question}")
    ]
)
```

### Initialize Output Parser

```python
from langchain_core.output_parsers import StrOutputParser

# Initialize the string output parser
parser = StrOutputParser()
```

```python
# Chain the prompt, LLM, and output parser
chain = prompt | llm | parser
```

```python
question = "My name is Sankajith Jinasena"
response = chain.invoke({"question":question})
print(response) # return a output
```

### Initialuze prompt Template for Dynamic Interaction

```python
from langchain_core.prompts import MessagesPlaceholder
from langchain_core.messages import HumanMessage, AIMessage, SystemMessage

prompt = ChatPromptTemplate.from_messages(
    [
        SystemMessage(content="You are an intelligent chatbot. Answer the follwong question"),
        MessagesPlaceholder(variable_name="question")
    ]
)

# Chain the prompt, LLM, and output parser
chain = prompt | llm | parser

question = "My name is Sankajith Jinasena"
response = chain.invoke({"question":HumanMessage(context=question)})
print(response) # return a output
```

### Initialize Prompt Template with Predefined Conversation History

```python
prompt = ChatPromptTemplate.from_messages(
    [
        SystemMessage(content="You are an intelligent chatbot. Answer the follwong question"),
        HumanMessage(content="My name is Sankajith Jinasena"),
        AIMessage(content="Nice to meet you, Sankajith Jinasena! How can I assist you today?"),
        MessagesPlaceholder(variable_name="question")
    ]
)

# Chain the prompt, LLM, and output parser
chain = prompt | llm | parser

question = "Who am I"
response = chain.invoke({"question":HumanMessage(context=question)})
print(response) # return a output
```

### Initialize Prompt Template to handle Dynamic Conversation History

```python
prompt = ChatPromptTemplate.from_messages(
    [
        SystemMessage(content="You are an intelligent chatbot. Answer the follwong question"),
        MessagesPlaceholder(varable_name="history"),
        MessagesPlaceholder(variable_name="question")
    ]
)

history = [
    HumanMessage(content="My name is Sankajith Jinasena"),
    AIMessage(content="Nice to meet you, Sankajith Jinasena! How can I assist you today?"),
    HumanMessage(content="what is 2+2"),
    AIMessage(content="4")
]

chain = prompt | llm | parser


question = "Who am I"
response = chain.invoke({"history":history,"question":HumanMessage(context=question)})
print(response)

# Update history
history.extend([HumanMessage(content=question),AIMessage(content=response)])

# We need to limit the history due to token limit
```
